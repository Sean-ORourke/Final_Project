We begin by loading in the data with pandas.

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Now we read the file containing the data

school_data = pd.read_csv("schooldata.csv")
# Ignore these comments for now
#plot = school_data["Estimate; Total:"].plot.bar(ascending = False)
#plot = school_data.sort_values(["Estimate; Total:"], ascending = False).set_index("Neighborhood")["Estimate; Total:"].head(10).plot.bar()
#school_data.head(3)

school_data.head(3)
```

The two columns to look at here are Neighborhood and Estimate; Enrolled in school:

The neighborhoods with the highest amounts of enrollment will be our contenders.

```python
# First let's sort the data in descending order by enrollment
plot = school_data.sort_values(["Estimate; Total:"], ascending = False)
plot.head(10)
```

The neighborhoods you see on the left are our top contenders.  Now let's try to get this in a graph.

```python
plot = school_data.sort_values(["Estimate; Total:"], ascending = False).set_index("Neighborhood")["Estimate; Total:"].plot.bar()
```

That's a lot of neighborhoods.  Let's just look at the top 20.

```python
plot = school_data.sort_values(["Estimate; Total:"], ascending = False).set_index("Neighborhood")["Estimate; Total:"].head(20).plot.bar()
plot.set_ylabel("Enrollment")
```

That's a lot better.  And it looks like our clear winner is Squirrel Hill South.
