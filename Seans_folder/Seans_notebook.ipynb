## Rate of Crime per Neighborhood in Pittsburgh

Here I am evaluating what is the best neighborhood in Pittsburgh for kids based on which neighborhoods have the least rate of crime, and the least rate of violent crime (crimes are evaluated based on how far left (violent) versus right (tame) they are on the below csv:

```python
# Importing pandas and getting a look at the csv
import pandas as pd
safe = pd.read_csv("public-safety.csv", sep=',')

safe.head()
```

After loading in the csv file with all the information, and taking a look at it; I plotted a bar chart the neighborhoods on the x axis, and the crime rate per 100 people on the y axis.

So, first we will look at the total reports of crime across all neighborhoods:

```python
ax = safe.plot.bar(x = "Neighborhood", y = "Part 1 Crime per 100 Persons (2010)", figsize = (15, 10), title = "Crime per Neighborhood")
ax.set_xlabel("Neighborhood")
ax.set_ylabel("Crime per 100 Persons")
```

Because this graph is hard to understand, I wanted to sort my data. I took those same values from the crime rate per 100 people y column, and sorted all the neighborhoods by that number, least to greatest.

Now we are sorting our neighborhoods least (head) to most (tail) crime:

```python
least_crime = safe.sort_values("Part 1 Crime per 100 Persons (2010)", ascending = True)

least_crime.head()
```

Then, I made another bar graph, getting a clearer look at the distribution of crime rates throughout the neighborhoods.

```python
ax2 = least_crime.plot.bar(x = "Neighborhood", y = "Part 1 Crime per 100 Persons (2010)", figsize = (15, 10), title = "Crime per Neighborhood, Sorted Least to Greatest")
ax2.set_xlabel("Neighborhood")
ax2.set_ylabel("Crime per 100 Persons")
```

I chose to display the last five neighborhoods in the sorted list, to get a closer look at the 5 most crime filled neighborhoods in Pittsburgh.

Now we are zooming in to look at the tail (most crime):

```python
ax4 = least_crime.tail(5).plot.bar(x = "Neighborhood", y = "Part 1 Crime per 100 Persons (2010)", title = "Most Crime per Neighborhood")
ax4.set_xlabel("Neighborhood")
ax4.set_ylabel("Crime per 100 Persons")
```

Then, I chose to display the first five neighborhoods in the sorted list, to get a closer look at the 5 most crime filled neighborhoods in Pittsburgh.

Now we are looking at head (least crime):

```python
ax3 = least_crime.head(5).plot.bar(x = "Neighborhood", y = "Part 1 Crime per 100 Persons (2010)", title = "Least Crime per Neighborhood")
ax3.set_xlabel("Neighborhood")
ax3.set_ylabel("Crime per 100 Persons")
```

But, not all crime is created equal. I added up each type of crime, chose a color for each category, and stacked them on top of eachother.

Neighborhoods with most crime reports by type:

```python
ax6 = least_crime.tail(5).plot.bar(x = "Neighborhood", y = ["#Murder (2010)", "#Rape (2010)", "#Robbery (2010)", "#Agr. Assault (2010)", "#Burglary (2010)", "#Auto Theft (2010)", "#Drug Violations (2010)"], stacked = True, color={"#Murder (2010)": "maroon", "#Rape (2010)": "red", "#Robbery (2010)": "slategray", "#Agr. Assault (2010)": "violet", "#Burglary (2010)": "gainsboro", "#Auto Theft (2010)": "cornflowerblue", "#Drug Violations (2010)": "gold"}, title = "Crime by type per Neighborhood with Most Crime")
ax6.set_xlabel("Neighborhood")
ax6.set_ylabel("Crime by Type")
```

Then I did this with the neighborhoods I sorted previously with the least crime.

Neighborhoods with least crime reports by type:

```python
ax5 = least_crime.head(5).plot.bar(x = "Neighborhood", y = ["#Murder (2010)", "#Rape (2010)", "#Robbery (2010)", "#Agr. Assault (2010)", "#Burglary (2010)", "#Auto Theft (2010)", "#Drug Violations (2010)"], stacked = True, color={"#Murder (2010)": "maroon", "#Rape (2010)": "red", "#Robbery (2010)": "slategray", "#Agr. Assault (2010)": "violet", "#Burglary (2010)": "gainsboro", "#Auto Theft (2010)": "cornflowerblue", "#Drug Violations (2010)": "gold"}, title = "Crime by type per Neighborhood with Least Crime")
ax5.set_xlabel("Neighborhood")
ax5.set_ylabel("Crime by Type")
```

Then, just for fun, I generated a last stacked bar graph with all neighborhoods and the types of crimes committed within them:

```python
ax7 = least_crime.plot.bar(x = "Neighborhood", y = ["#Murder (2010)", "#Rape (2010)", "#Robbery (2010)", "#Agr. Assault (2010)", "#Burglary (2010)", "#Auto Theft (2010)", "#Drug Violations (2010)"], stacked = True, color={"#Murder (2010)": "maroon", "#Rape (2010)": "red", "#Robbery (2010)": "slategray", "#Agr. Assault (2010)": "violet", "#Burglary (2010)": "gainsboro", "#Auto Theft (2010)": "cornflowerblue", "#Drug Violations (2010)": "gold"}, figsize = (15, 10), title = "Crime by type per Neighborhood")
ax7.set_xlabel("Neighborhood")
ax7.set_ylabel("Crime by Type")
```
